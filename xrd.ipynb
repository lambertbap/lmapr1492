```python
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from matplotlib import pyplot as plt
from tabulate import tabulate

with MPRester("KwnCeSBHTUbZ9VP0") as m:
    struc=m.get_structure_by_material_id("mp-1018100")
    CifWriter(struc).write_file("AlSb.cif")
    analyzer=SpacegroupAnalyzer(struc).get_primitive_standard_structure()
xrd = XRDCalculator(wavelength = 1.54060)
motif_xrd= xrd.get_pattern(analyzer,scaled=True)
p=0
q=1
while p<len([0,1,2]):
    print(q,"e pic de diffraction :")
    print(motif_xrd.hkls[p][0]['hkl'])
    print("Observé en un angle de 2θ =", motif_xrd.x[p],"°")
    p+=1
    q+=1
xrd.show_plot(analyzer)
```
