```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pprint import pprint
from pymatgen.io.cif import CifWriter
from pymatgen import Structure
from pymatgen import symmetry
from pymatgen.core.operations import SymmOp
from IPython.display import Image


with MPRester("KwnCeSBHTUbZ9VP0") as m:
    # Structure for material id
    struc = m.get_structure_by_material_id("mp-1018100",False)
    S = SpacegroupAnalyzer(struc)
    CifWriter(struct=struc, symprec=True).write_file("AlSb.cif")
    
    
print("2e opération de sym, 6-fold screw axis translation:")
Opp2=SymmOp.from_xyz_string("x-y,x,z+1/2")
print(Opp2)
print("")
Ef2=Opp2.operate((0,0,1/2))
print(Ef2)
Image("2e.PNG")
```

```python
print("3e opération de sym, c3 axis:")
Opp3=SymmOp.from_xyz_string("-y,x-y,z")
print(Opp3)
print("")
Ef3=Opp3.operate((0,0,0))
print(Ef3)
Image("3e.PNG")
```

```python
print("4e opération de sym, 2-fold screw axis translation:")
Opp4=SymmOp.from_xyz_string("-x,-y,z+1/2")
print(Opp4)
print("Effet sur Tb1 #2:")
Ef4=Opp4.operate((0,0,1/2))
print(Ef4)
Image("4e.PNG")
```
